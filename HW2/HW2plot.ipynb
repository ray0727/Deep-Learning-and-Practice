```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(np.arange(100000))
```

```python
def plot(num):
    plt.figure(figsize=(10,5)) 
    plt.title("score figure", fontsize=18)
    plt.plot(np.arange(100001), num)
    my_x_ticks = np.arange(0, 100001, 20000)
    my_y_ticks = np.arange(0, 160001, 10000)
    plt.xticks(my_x_ticks)
    plt.yticks(my_y_ticks)
    plt.ylabel("score", fontsize=16)
    plt.xlabel("episodes", fontsize=16)
    plt.show()
```

```python
if __name__ == "__main__":
    num = []
    num.append(0)
    f = open("/content/score.txt")
    lines = f.readlines()
    for l in lines[0:len(lines)]:
        num.append(int(l))
    f.close()
    plot(num)
```

```python
max = []
max_ep = 0
max.append(0)
for i in range(len(num)):
  if(max_ep<num[i]):
    max_ep = num[i]
  if((i+1)%1000 == 0):
    max.append(max_ep)
    max_ep = 0
```

```python
print(max)
print(len(max))
```

```python
def plot_max(num):
    plt.figure(figsize=(10,5)) 
    plt.title("score max per 1000 epsiode", fontsize=18)
    plt.plot(np.arange(101), num)
    my_x_ticks = np.arange(0, 101, 10)
    my_y_ticks = np.arange(0, 160000, 10000)
    plt.xticks(my_x_ticks)
    plt.yticks(my_y_ticks)
    plt.ylabel("score", fontsize=16)
    plt.xlabel("1000 episodes", fontsize=16)
    plt.show()
```

```python
plot_max(max)
```
